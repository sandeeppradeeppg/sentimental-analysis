```python
def rating(rating):
  if rating > 3 and rating <= 5:
      return "Postive"
  if rating > 0 and rating <= 3:
      return "Negative"
data["Rating_posorneg"] = data["Rating"].apply(rating)
```

Sentiment analysis

```python
import nltk
```

```python
nltk.download("vader_lexicon")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sentanalyze = SentimentIntensityAnalyzer()
```

```python
data["Scores"] = data["Review"].apply(lambda review: sentanalyze.polarity_scores(review))
```

```python
data.head()
```

```python
data["compound"] = data["Scores"].apply(lambda f:f["compound"])
```

```python
data.head()
```

```python
data["compund_posorneg"] = data["compound"].apply(lambda score:"Positive" if score >= 0 else "Negative")
```

```python
data.head()
```

```python
example = "Hotel was excellent and i Enjoyed it"
sentanalyze.polarity_scores(example)
```

--classification of reviews

```python
from sklearn.model_selection import train_test_split
```

```python
x = data["Review"]
y = data["Rating_posorneg"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.svm import LinearSVC
```

```python
textclf = Pipeline([("tfidf",TfidfVectorizer()),("clf",LinearSVC())])
```

```python
textclf.fit(x_train,y_train)
```

```python
predictions = textclf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,predictions))
```
